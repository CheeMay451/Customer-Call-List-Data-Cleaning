```python
# drop the duplicates

df = df.drop_duplicates()
df
```

```python
# remove any unecessary columns

df = df.drop(columns = 'Not_Useful_Column')
df
```

```python
# Specify column 'Last_Name' and trim the last name

#df['Last_Name'] = df['Last_Name'].str.lstrip('/')
#df['Last_Name'] = df['Last_Name'].str.lstrip('.')
#df['Last_Name'] = df['Last_Name'].str.rstrip('_')
#df
```

```python
# Alternatively, use this method:

df['Last_Name'] = df['Last_Name'].str.strip('123./_')
df
```

```python
# Specify column 'Phone_Number'  find characters that are NOT letters or numbers  deletes it

df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]', '', regex = True) # Without regex=True, it may not recognize the pattern as a regex expression.
df
```

```python
# 'Phone_Number' includes float and string
# Convert it to string before formatting it

df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
```

```python
# Format the 'Phone_Number'

df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
```

```python
df
```

```python
# Removes NaN

df['Phone_Number'] = df['Phone_Number'].str.replace('nan--', '')

df['Phone_Number'] = df['Phone_Number'].str.replace('Na--', '')

df
```

```python
# splitting columns

df[['Street', 'State', 'Zip_Code']]= df['Address'].str.split(',', n = 2, expand = True)
df

# Use double square brackets for selecting or assigning multiple columns at once
# Split the 'Address' column into multiple columns using a comma as the separator
# 'n=2' ensures at most 2 splits, creating three columns 
# 'expand=True' ensures the split results are returned as separate columns instead of lists
```

```python
# Standardising Column Values

df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y')
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')

df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y')
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No', 'N')

df
```

```python
# Remove NULL values

df = df.replace('NaN', '')
df = df.replace('N/a', '')

df
```

```python
# Some of the Null values are not removable because they are technically empty. Had to fill NA with space.

df = df.fillna('')
df
```

```python
# Remove rows of data of participants who specified not to contact them

for row_data in df.index:
    if df.loc[row_data, 'Do_Not_Contact'] == 'Y':
        df.drop(row_data, inplace = True)
```

```python
df
```

```python
for row_data in df.index:
    if df.loc[row_data, 'Phone_Number'] == '':
        df.drop(row_data, inplace = True)

```

```python
df
```

```python
# reset index

df = df.reset_index(drop = True)
```
